```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

```python
df_1=pd.read_csv("train.csv")
df_1
```

```python
df_2=pd.read_csv("test.csv")
df_2
```

```python
df_3=pd.concat([df_1,df_2])
df_3
```

```python
X=df_3.drop('Onshore/Offshore',axis=1)
y=df_3['Onshore/Offshore']
```

```python
X_transform=pd.get_dummies(X)
X_transform
```

```python
df=pd.concat([X_transform,y],axis=1)
df
```

```python
df_train = df.dropna()
df_test= df[df['Onshore/Offshore'].isnull()]
df_test = df_test.drop('Onshore/Offshore', axis = 1)
```

```python
df_train['Onshore/Offshore'].unique()
```

```python
df_train['Onshore/Offshore']=df_train['Onshore/Offshore'].astype('category')
```

```python
df_y=df_train['Onshore/Offshore'].cat.codes
df_y
```

```python
dict_shore={0:'OFFSHORE',
              1:'ONSHORE',
              2:'ONSHORE-OFFSHORE'}
```

```python
X=df_train.drop('Onshore/Offshore', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, df_y, test_size=0.3, random_state=42)
y_train.value_counts()
```

```python
clf=OneVsRestClassifier(SVC()).fit(X_train, y_train)
```

```python
y_predict=clf.predict(X_test)
```

```python
print(classification_report(y_test,y_predict))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
scale=StandardScaler()
X_scale=scale.fit_transform(X)
rfc_1 = RandomForestClassifier()
rfc_1.fit(X_train, y_train)
y_pred_1 = rfc_1.predict(X_test)
print(classification_report(y_test,y_pred_1))
```
